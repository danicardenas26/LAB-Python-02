```python

# Ejercicio 5 - Combinación de funciones

# Datos de prueba
numeros = [2, 5, 8, 11, 15, 3, 20]

# a. Filtrar números pares y luego elevarlos al cuadrado.
pares_al_cuadrado = list(map(lambda x: x**2, filter(lambda x: x % 2 == 0, numeros)))

# b. Dada una lista de numeros, elevar cada número al cubo, filtrar los mayores que 100, convertirlos a cadenas.
# Se lee de adentro hacia afuera:
# 1. map (elevar al cubo)
# 2. filter (quedarse con los > 100)
# 3. map (convertir a string con str)
cubos_mayores_100_str = list(map(str, filter(lambda x: x > 100, map(lambda x: x**3, numeros))))

# c. Dada una lista de numeros, filtrar números mayores que 10, elevarlos al cuadrado, ordenarlos de mayor a menor.
# 1. filter (mayores que 10)
# 2. map (elevar al cuadrado)
# 3. sorted (con reverse=True para que sea de mayor a menor)
mayores_10_cuadrados_desc = sorted(map(lambda x: x**2, filter(lambda x: x > 10, numeros)), reverse=True)


# Código para imprimir y comprobar los resultados
print("Lista original:", numeros)
print("a. Pares al cuadrado:", pares_al_cuadrado)
print("b. Cubos > 100 como texto:", cubos_mayores_100_str)
print("c. > 10 al cuadrado (descendente):", mayores_10_cuadrados_desc)
```
