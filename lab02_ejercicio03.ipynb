```python
import math # Importamos math para usar el valor exacto de pi en el cálculo del área

# Datos de prueba
temperaturas_celsius = [0, 20, 30, 100]
lista_a = [1, 2, 3, 4]
lista_b = [10, 20, 30, 40]
radios = [1, 2.5, 4, 10]
productos = {"pan": 1000, "leche": 2500, "café": 5000}
matriz = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]


# a. Dada una lista de temperaturas en Celsius, convertirlas a Fahrenheit.
temp_fahrenheit = list(map(lambda c: (c * 9/5) + 32, temperaturas_celsius))

# b. Dadas dos listas con la misma cantidad de elementos, generar una nueva sumando cada elementos.
suma_listas = list(map(lambda x, y: x + y, lista_a, lista_b))

# c. Dada una lista de radios de circulos, generar la lista de sus areas.
areas_circulos = list(map(lambda r: math.pi * (r ** 2), radios))

# d. Dada un diccionario, generar una lista de tuplas (nombre, valor+10%)
precios_con_aumento = list(map(lambda item: (item[0], item[1] * 1.10), productos.items()))

# e. Dada una matriz, multiplicar todos los elementos por 10.
matriz_por_10 = list(map(lambda fila: list(map(lambda x: x * 10, fila)), matriz))



print("a.", temp_fahrenheit)
print("b.", suma_listas)
# Redondeamos a 2 decimales
print("c.", [round(area, 2) for area in areas_circulos])
print("d.", precios_con_aumento)
print("e.", matriz_por_10)
```
