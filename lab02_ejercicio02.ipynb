```python
# Datos de prueba
numeros = [5, 8, 12, 15, 3, 20]
temperaturas_celsius = [0, 20, 30, 100]
palabras = ["manzana", "sol", "programacion", "hola", "universidad"]
matriz = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]

# a. Generar una lista con tuplas (n, n^2) para una lista de numeros.
tuplas_cuadrados = [(n, n**2) for n in numeros]

# b. Generar una lista con cubos de los números mayores a 10.
cubos_mayores_10 = [n**3 for n in numeros if n > 10]

# c. Generar una lista con solo los numeros pares.
numeros_pares = [n for n in numeros if n % 2 == 0]

# d. Dada la lista numeros, producir: ["valor: 10", "valor: 15", ...]
lista_valores = [f"valor: {n}" for n in numeros]

# e. Generar una lista con la conversion correspondiente a grados Fahrenheit.
temperaturas_fahrenheit = [(c * 9/5) + 32 for c in temperaturas_celsius]

# f. Generar una lista filtrando las palabras de mas de 5 letras.
palabras_mas_de_5 = [palabra for palabra in palabras if len(palabra) > 5]

# g. Generar una lista con solo las iniciales de las palabras.
iniciales = [palabra[0] for palabra in palabras]

# h. Generar una lista indicando "par" o "impar" dada una lista de numeros.
pares_e_impares = ["par" if n % 2 == 0 else "impar" for n in numeros]

# i. Dada una lista de listas (matriz), aplanala.
lista_aplanada = [elemento for sublista in matriz for elemento in sublista]


print("a.", tuplas_cuadrados)
print("b.", cubos_mayores_10)
print("c.", numeros_pares)
print("d.", lista_valores)
print("e.", temperaturas_fahrenheit)
print("f.", palabras_mas_de_5)
print("g.", iniciales)
print("h.", pares_e_impares)
print("i.", lista_aplanada)
```
