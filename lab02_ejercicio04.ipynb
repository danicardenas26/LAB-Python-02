```python
from functools import reduce

# Ejercicio 4 - Filter

# Datos de prueba
palabras_f = ["oso", "programacion", "elefante", "ana", "sol", "iglesia"]
lista_nulos = [1, None, 3, None, "hola", False]
numeros_f = [10, 15, 20, 25, 30]
productos_f = [
    {"nombre": "Teclado", "precio": 80},
    {"nombre": "Mouse", "precio": 40},
    {"nombre": "Monitor", "precio": 300},
    {"nombre": "Webcam", "precio": 150}
]

# a. Dada una lista de palabras, filtrar solo aquellas con mas de 4 letras.
mas_de_4_letras = list(filter(lambda p: len(p) > 4, palabras_f))

# b. Dada una lista filtrar los elementos que son nulos (None).

solo_nulos = list(filter(lambda x: x is None, lista_nulos))

# c. Dada una lista de palabras, filtrar aquellas que empiezan por una vocal.
empiezan_vocal = list(filter(lambda p: p[0].lower() in 'aeiou', palabras_f))

# d. Dada una lista de palabras, filtrar aquellas que son palindromos.
palindromos = list(filter(lambda p: p.lower() == p[::-1].lower(), palabras_f))

# e. Dada una lista de numeros, filtrar los que terminan en 5.
terminan_en_5 = list(filter(lambda n: n % 10 == 5, numeros_f))

# f. Dada una lista de diccionarios, filtrar los productos con precio mayor a 100.
precio_mayor_100 = list(filter(lambda prod: prod["precio"] > 100, productos_f))

# Ejercicio 4 - Reduce

# Datos de prueba
numeros_r = [2, 3, 4, 5]
palabras_r = ["Hola", " ", "mundo", "!"]

# a. Dada una lista de numeros, calcula la multiplicación de estos.
multiplicacion_total = reduce(lambda x, y: x * y, numeros_r)

# b. Dada una lista de palabras, concatenalas en una sola.
palabras_concatenadas = reduce(lambda x, y: x + y, palabras_r)

# c. Dada una lista, hallar el mayor de todos, define una función auxiliar...
def mayor_de_dos(x, y):
    return x if x > y else y

numero_mayor = reduce(mayor_de_dos, numeros_r)

# Ejercicio 4 - Sorted

# Datos de prueba
palabras_s = ["zapato", "arbol", "casa", "murcielago"]
tuplas_s = [(3, 'c'), (1, 'a'), (2, 'b')]
productos_s = [
    ("Camisa", 30),
    ("Pantalón", 55),
    ("Medias", 10),
    ("Chaqueta", 80)
]
numeros_s = [4, 1, 8, 3, 2, 7]
empleados = [
    ("Maria", "Ventas", 30),
    ("Luis", "Tecnología", 25),
    ("Ana", "Ventas", 25),
    ("Pedro", "Tecnología", 28)
]

# a. Ordenar una lista de palabras alfabeticamente.
palabras_alfabetico = sorted(palabras_s)

# b. Ordenar una lista de tuplas por el primer elemento.
tuplas_ordenadas = sorted(tuplas_s, key=lambda x: x[0])

# c. Dada una lista de diccionarios (usaremos las tuplas de la imagen), ordenar por precio (mayor a menor), y luego por nombre.
productos_ordenados = sorted(productos_s, key=lambda x: (-x[1], x[0]))

# d. Dada una lista de palabras, ordenarlas según su longitud.
palabras_por_longitud = sorted(palabras_s, key=len)

# e. Dada una lista de numeros, ordenarlos de tal forma que aparezcan primero los pares y luego los impares.
numeros_pares_impares = sorted(numeros_s, key=lambda x: (x % 2, x))

# f. Convertir una palabra a un diccionario que cuente la aparicion de cada letra, ordenar dicho diccionario.
palabra_f = "programacion"
diccionario_letras = {letra: palabra_f.count(letra) for letra in set(palabra_f)}
diccionario_ordenado = dict(sorted(diccionario_letras.items()))

# g. Dada una lista de tuplas, ordenar la primero por el departamento del empleado y luego por la edad.
empleados_ordenados = sorted(empleados, key=lambda x: (x[1], x[2]))


# Impresiones de comprobación rápida
print("--- Resultados Filter ---")
print("d. Palíndromos:", palindromos)
print("f. Precio > 100:", precio_mayor_100)

print("\n--- Resultados Reduce ---")
print("a. Multiplicación:", multiplicacion_total)
print("c. Mayor:", numero_mayor)

print("\n--- Resultados Sorted ---")
print("c. Productos ordenados:", productos_ordenados)
print("e. Pares luego impares:", numeros_pares_impares)
print("g. Empleados ordenados:", empleados_ordenados)
```
